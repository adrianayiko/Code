```python
import requests
from bs4 import BeautifulSoup
r = requests.get('https://www.emirates.com/ug/English/destinations/flights-from-entebbe.aspx')
soup = BeautifulSoup(r.text, 'lxml')
print (soup.title)
print (soup.title.text)
```

```python
soup1 = BeautifulSoup(r.content)
#print html of url
print(soup1.prettify())
```

```python
for link in soup1.find_all('link'):
    print(link.get('href'))
```
