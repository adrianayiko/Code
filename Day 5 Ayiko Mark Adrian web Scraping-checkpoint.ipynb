```python
#web scraping from  https://xeno-canto.org/
import requests
url ='https://www.xeno-canto.org/api/2/recordings?query=cnt:uganda'
headers = {
    'Accept': 'application/json',
    'Content-Type': 'application/json'
}
response = requests.request('GET',url,headers=headers,data={})
myjson = response.json()
import pandas as pd
recordings = myjson['recordings']
recording = pd.DataFrame(data=recordings)
recording.to_csv('recording_UG.csv',index=False)
```

```python
uganda = pd.read_csv('recording_UG.csv')
print(uganda.to_string())
```

```python
#Web scripping using beautiful soup
import requests
from bs4 import BeautifulSoup as bs
```

```python
request = requests.get('https://keithgalli.github.io/web-scraping/webpage.html')
#convert to beautiful soup page
soup = bs(request.content)
#print html of url
print(soup.prettify())
```
